## 8. Data-Cleaning

```python
# Getting columns where at least one '?' exists
columns_ques = df_income.columns[df_income.eq('?').any()].tolist()

# Getting the count of '?' per column
ques_count = df_income.eq('?').sum()[columns_ques]

print("Columns containing '?':", columns_ques)
print("\nCount of '?' in each column:")
print(ques_count)
```

```python
# Replacing "?" with NaN
df_income.replace(' ?', np.nan, inplace=True)
```

```python
# Dropping rows with missing values
df_income.dropna(inplace=True)
```

## 9. Feature and Target Separation

```python
# Separating features and target (income)
# Converting the values in income column to binary class
# >50K -> 1
# <=50K -> 0
X = df_income.drop('income', axis=1)
y = df_income['income'].apply(lambda x: 1 if x.strip() == '>50K' else 0)
```

## 10. Feature Engineering

```python
# Selecting categorical and numerical columns
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()
numerical_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Defined the transformers
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

categoric_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])


# Combining the transformers
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_cols),
        ('cat', categoric_transformer, categorical_cols)
    ])

# Applying transformations
X_engg = preprocessor.fit_transform(X)
```

## 11. Train-Test Splitting

```python
# Splitting into train and test
X_train, X_test, y_train, y_test = train_test_split(X_engg, y, test_size = 0.2, random_state = 12)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Training labels distribution:\n", y_train.value_counts(normalize = True))
```

## 12. Building and compiling the model

```python
# Build the NN model
NN_4_model = Sequential([
    Input(shape=(X_train.shape[1],)), # Explicit Input layer
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')  # Output layer for binary classification
])

# Compile the model
NN_4_model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

## 13. Training the model

```python
train_hist = NN_4_model.fit(X_train, y_train,
                    epochs = 30,
                    batch_size = 64,
                    validation_split = 0.2,
                    callbacks = [EarlyStopping(patience = 3, restore_best_weights = True)],
                    verbose = 2)
```

## 14. Evaluating the model

```python
# Evaluate on test set
test_loss, test_acc = NN_4_model.evaluate(X_test, y_test)
print(f'\nTest accuracy: {test_acc:.4f}')
```

## 15. Plot model accuracy

```python
# Plot Accuracy
plt.figure(figsize = (10,6))
plt.plot(train_hist.history['accuracy'], label = 'Train Accuracy')
plt.plot(train_hist.history['val_accuracy'], label = 'Validation Accuracy')
plt.title('Training vs Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

## 16. Plot model loss

```python
# Plot Loss
plt.figure(figsize = (10,6))
plt.plot(train_hist.history['loss'], label = 'Train Loss')
plt.plot(train_hist.history['val_loss'], label = 'Validation Loss')
plt.title('Training vs Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

## 17. Generating predictions

```python
y_predictions = (NN_4_model.predict(X_test) > 0.5).astype(int)
```

## 18. Classification Report

```python
print(classification_report(y_test, y_predictions))
```
