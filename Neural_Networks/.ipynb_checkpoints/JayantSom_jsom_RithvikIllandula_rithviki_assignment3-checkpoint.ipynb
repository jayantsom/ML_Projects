## 1. Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.callbacks import EarlyStopping
```

## 8. Data-Cleaning

```python
# Getting columns where at least one '?' exists
columns_ques = df_income.columns[df_income.eq('?').any()].tolist()

# Getting the count of '?' per column
ques_count = df_income.eq('?').sum()[columns_ques]

print("Columns containing '?':", columns_ques)
print("\nCount of '?' in each column:")
print(ques_count)
```

```python
# Replacing "?" with NaN
df_income.replace(' ?', np.nan, inplace=True)
```

```python
# Dropping rows with missing values
df_income.dropna(inplace=True)
```

## 9. Feature and Target Separation

```python
# Separating features and target (income)
# Converting the values in income column to binary class
# >50K -> 1
# <=50K -> 0
X = df_income.drop('income', axis=1)
y = df_income['income'].apply(lambda x: 1 if x.strip() == '>50K' else 0)
```

## 10. Feature Engineering

```python
# Selecting categorical and numerical columns
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()
numerical_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Defined the transformers
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

categoric_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])


# Combining the transformers
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_cols),
        ('cat', categoric_transformer, categorical_cols)
    ])

# Applying transformations
X_engg = preprocessor.fit_transform(X)
```

## 11. Train-Test Splitting

```python
# Splitting into train and test
X_train, X_test, y_train, y_test = train_test_split(X_engg, y, test_size = 0.2, random_state = 12)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Training labels distribution:\n", y_train.value_counts(normalize = True))
```

## 12. Building and compiling the model

```python
# Build the NN model
NN_4_model = Sequential([
    Input(shape=(X_train.shape[1],)), # Explicit Input layer
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')  # Output layer for binary classification
])

# Compile the model
NN_4_model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

## 13. Training the model

```python
train_hist = NN_4_model.fit(X_train, y_train,
                    epochs = 30,
                    batch_size = 64,
                    validation_split = 0.2,
                    callbacks = [EarlyStopping(patience = 3, restore_best_weights = True)],
                    verbose = 2)
```

## 14. Evaluating the model

```python
# Evaluate on test set
test_loss, test_acc = NN_4_model.evaluate(X_test, y_test)
print(f'\nTest accuracy: {test_acc:.4f}')
```

## 15. Plot model accuracy

```python
# Plot Accuracy
plt.figure(figsize = (10,6))
plt.plot(train_hist.history['accuracy'], label = 'Train Accuracy')
plt.plot(train_hist.history['val_accuracy'], label = 'Validation Accuracy')
plt.title('Training vs Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

## 16. Plot model loss

```python
# Plot Loss
plt.figure(figsize = (10,6))
plt.plot(train_hist.history['loss'], label = 'Train Loss')
plt.plot(train_hist.history['val_loss'], label = 'Validation Loss')
plt.title('Training vs Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

## 17. Generating predictions

```python
y_predictions = (NN_4_model.predict(X_test) > 0.5).astype(int)
```

## 18. Classification Report

```python
print(classification_report(y_test, y_predictions))
```

## 1. Imports

```python
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense, Input
from tensorflow.keras.callbacks import EarlyStopping
```

## 2. Data Loading

```python
# Loaded the Fashion MNIST dataset
(trainX, trainy), (testX, testy) = fashion_mnist.load_data()
```

## 3. Dataset Details

```python
# Showing dataset structure
print("Training data shape:", trainX.shape)
print("Test data shape:", testX.shape)
print("Number of classes:", len(np.unique(trainy)))
```

```python
# Class labels
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
# Show class distribution
unique, count = np.unique(trainy, return_counts=True)
print("Class distribution in training set:")
for label, cnt in zip(unique, count):
    print(f"{class_names[label]} ({label}): {cnt} samples")
```

## 4. Visualization

```python
# Visualizing 25 samples from the dataset
plt.figure(figsize = (10,10))

for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(trainX[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[trainy[i]])
    
plt.show()
```

## 5. Data Preprocessing

```python
# Normalizing pixel values to [0, 1]
trainX = trainX.astype('float32') / 255.0
testX = testX.astype('float32') / 255.0
```

```python
# Reshaping to include channel dimension for CNN input
trainX = trainX.reshape(-1, 28, 28, 1)
testX = testX.reshape(-1, 28, 28, 1)
```

```python
# One-hot encoding of the labels
trainy_cat = to_categorical(trainy, num_classes=10)
testy_cat = to_categorical(testy, num_classes=10)
```

```python
# Printing modified shape
print("trainX shape:", trainX.shape)
print("trainy_cat shape:", trainy_cat.shape)
print("testX shape:", testX.shape)
print("testy_cat shape:", testy_cat.shape)
```

## 6. Creating CNN Model

```python
# Defining the CNN model
model_4_CNN = Sequential([
    
    Input(shape=(28, 28, 1)), # Input layer
    
    Conv2D(32, (3, 3), activation = 'relu', padding = 'same'),
    MaxPooling2D(pool_size = (2, 2)),
    
    Conv2D(64, (3, 3), activation = 'relu'),
    MaxPooling2D(pool_size = (2, 2)),
    Dropout(0.25),
    
    Flatten(),
    
    Dense(128, activation='relu'),
    Dropout(0.5),
    
    Dense(10, activation='softmax')  # 10 classes output
])
```

```python
# Compiling the model
model_4_CNN.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
# Printing model summary
model_4_CNN.summary()
```

## 7. Training the model

```python
train_hist_cnn = model_4_CNN.fit(trainX, trainy_cat,
                    epochs = 20,
                    batch_size = 128,
                    validation_split = 0.2,
                    callbacks = [EarlyStopping(patience = 4, restore_best_weights = True)])
```

## 8. Model evaluation

```python
# Evaluating on test set
test_loss_cnn, test_acc_cnn = model_4_CNN.evaluate(testX, testy_cat, verbose = 0)
print(f"\n Test Accuracy: {test_acc_cnn:.4f}")
```

## 9. Plot of model accuracy

```python
plt.figure(figsize = (10, 5))

plt.plot(train_hist_cnn.history['accuracy'], label = 'Train Accuracy')
plt.plot(train_hist_cnn.history['val_accuracy'], label = 'Validation Accuracy')
plt.title('Training vs Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')

plt.legend()
plt.grid(True)
plt.show()
```

## 10. Plot of model loss

```python
plt.figure(figsize=(10, 5))

plt.plot(train_hist_cnn.history['loss'], label = 'Train Loss')
plt.plot(train_hist_cnn.history['val_loss'], label = 'Validation Loss')
plt.title('Training vs Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')

plt.legend()
plt.grid(True)
plt.show()
```

## 1. Imports

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau
```

## 2. Optimization technique 1 : Enhanced Early Stopping

```python
enhanced_early_stoppage = EarlyStopping(
    monitor = 'val_loss',
    patience = 8,
    restore_best_weights = True,
    min_delta = 0.001
)
```

## 3. Optimization technique 2 : Learning Rate Reduction

```python
reduced_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=3,
    min_lr=1e-6,
    verbose=1
)
```

## 4. Retraining the model with optimization

```python
train_hist_cnn_optimized = model_4_CNN.fit(
    trainX, trainy_cat,
    epochs = 30,  # Increased since we have early stopping
    validation_data = (testX, testy_cat),
    callbacks = [enhanced_early_stoppage, reduced_lr],
    verbose = 1
)
```

## 5. Evaluating the optimized model

```python
test_loss_cnn_opt, test_acc_cnn_opt = model_4_CNN.evaluate(testX, testy_cat, verbose = 0)

print(f"\nOriginal Test Accuracy: {test_acc_cnn:.4f}")
print(f"Optimized Test Accuracy: {test_acc_cnn_opt:.4f}")
```

## 6. Plot of model accuracy

```python
plt.figure(figsize = (10, 5))

plt.plot(train_hist_cnn_optimized.history['accuracy'], label = 'Train Accuracy')
plt.plot(train_hist_cnn_optimized.history['val_accuracy'], label = 'Validation Accuracy')
plt.title('Training vs Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')

plt.legend()
plt.grid(True)
plt.show()
```

## 7. Plot of model loss

```python
plt.figure(figsize=(10, 5))

plt.plot(train_hist_cnn_optimized.history['loss'], label = 'Train Loss')
plt.plot(train_hist_cnn_optimized.history['val_loss'], label = 'Validation Loss')
plt.title('Training vs Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')

plt.legend()
plt.grid(True)
plt.show()
```
